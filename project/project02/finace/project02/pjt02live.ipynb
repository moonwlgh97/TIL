### B데이터 전처리 - 2021 이후의 종가 데이터 출력하기

```python

#CSV 파일 경로
csv_path = 'data/NFLX.csv'

# CSV 파일 읽어오기 
new_df = pd.read_csv(csv_path, usecols=range(0,5))
# 2021년 이후 데이터 필터링
new_df['Date'] = pd.to_datetime(new_df['Date'])
df_after_2021 = new_df.loc[new_df['Date']>='2021-01-01']




#그래프 그리기 
plt.plot(df_after_2021['Date'], df_after_2021['Close'])

# 그래프 제목 설정
plt.title('NFLX Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

# y축 레이블 설정
plt.ylabel('Close')

# 그래프 표시
plt.show()
```

### C. 데이터 분석 - 2021 이후 최고 최저 종가 출력하기

```python
close_max_price=max(df_after_2021['Close'])
close_min_price=min(df_after_2021['Close'])

print('최고 종가 : ', close_max_price)
print('최저 종가 : ', close_min_price)
```

### D.데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기


```python
#CSV 파일 경로
csv_path = 'data/NFLX.csv'

# CSV 파일 읽어오기 
new_df = pd.read_csv(csv_path, usecols=range(0,5))
# 2021년 이후 데이터 필터링
new_df['Date'] = pd.to_datetime(new_df['Date'])
df_after_2021 = new_df.loc[new_df['Date']>='2021-01-01']


avr_df = df_after_2021.groupby(pd.Grouper(key='Date', freq='M')).mean().reset_index()
avr_df1 = pd.DataFrame(avr_df)



plt.plot(avr_df1['Date'], avr_df1['Close'])

# 그래프 제목 설정
plt.title('Monthly Average Close Price')

# x축 레이블 설정
plt.xlabel('Date')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

# y축 레이블 설정
plt.ylabel('Average Close Price')

# 그래프 표시
plt.show()

```

### E. 데이터 시각화 - 2022년 이후 최고, 최저 종가

```python
# CSV 파일 경로
csv_path = 'data/NFLX.csv'

# CSV 파일 읽어오기 (첫 번째, 마지막 열 제외)
new_df = pd.read_csv(csv_path, usecols=range(0, 5))
new_df['Date'] = pd.to_datetime(new_df['Date'])
# 2022년 이후 데이터 필터링
df_after_2022 = new_df[new_df["Date"] >= "2022-01-01"]

# 그래프 그리기
plt.plot(df_after_2022['Date'], df_after_2022['High'], label='High')
plt.plot(df_after_2022['Date'], df_after_2022['Low'], label='Low')
plt.plot(df_after_2022['Date'], df_after_2022['Close'], label='Close')

# 그래프 제목 설정
plt.title('High, Low, and Close Prices since January 2022')

# x축 레이블 설정
plt.xlabel('Date')

# x 축 설정(회전시키기)
plt.xticks(rotation=45)

# y축 레이블 설정
plt.ylabel('Price')

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```
